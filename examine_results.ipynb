## Sort and modify results log

```python
def add_results_to_record(model_name, sim_type, val_set_name, simulation_folder_path, log_parquet_path = './all_results.parquet'):
    sheet_name = model_name + '--' + sim_type + '--' + val_set_name
    input_parquet_path = simulation_folder_path + sim_type + '/' + val_set_name + '/aggregator_metric/'
    # read the only file in the directory
    input_parquet_path = input_parquet_path + os.listdir(input_parquet_path)[0]

    # copy the input parquet file to the results folder
    import shutil
    shutil.copy(input_parquet_path, os.path.expanduser('~/Documents/results/' + sheet_name + '.parquet'))
    # Read the input parquet file
    new_data = pd.read_parquet(input_parquet_path)
    
    # Check if the log parquet file exists
    if os.path.exists(log_parquet_path):
        # Read the existing log parquet file
        log_data = pd.read_parquet(log_parquet_path)
    else:
        raise FileNotFoundError('The log parquet file does not exist.')
    
    score_row = new_data.loc[new_data['scenario'] == 'final_score']
    score_row.insert(0, 'model', model_name)
    score_row.insert(1, 'simulation', sim_type)
    score_row.insert(2, 'validation', val_set_name)    
    
    log_data = pd.concat([log_data, score_row], ignore_index=True)
    # Write the log parquet file
    # log_data.to_parquet(log_parquet_path)
    return log_data

```

```python
# Example usage
model_name = 'best_1'
val_set_name = 'val14' # 'test14-random', 'test14-hard', 'val14'
# for sim_type in ['closed_loop_nonreactive_agents']:
for sim_type in ['closed_loop_nonreactive_agents', 'closed_loop_reactive_agents', 'open_loop_boxes']:
    # input_parquet_path = '/home/jiale/Documents/exp/exp/simulation/' + sim_type + '/' + val_set_name + '/aggregator_metric/'
    simulation_folder_path = '/home/jiale/Documents/exp/exp/simulation/'
    log_parquet_path = './all_results.parquet'
    log_data = add_results_to_record(model_name, sim_type, val_set_name, simulation_folder_path)
    log_data.to_parquet(log_parquet_path)
```
